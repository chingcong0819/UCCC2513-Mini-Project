```python
import numpy as np
import matplotlib.pyplot as plt

# Define dimensions
width, height = 256, 256

# Generate random noise for color image (3 channels: RGB)
color_noise = np.random.rand(height, width, 3)

# Generate random noise for grayscale image (1 channel)
grayscale_noise = np.random.rand(height, width)

# Plot the images
plt.figure(figsize=(10, 5))

# Color image
plt.subplot(1, 2, 1)
plt.imshow(color_noise)
plt.title('Random Noise Color Image')
plt.axis('off')

# Grayscale image
plt.subplot(1, 2, 2)
plt.imshow(grayscale_noise, cmap='gray')
plt.title('Random Noise Grayscale Image')
plt.axis('off')

plt.show()
```

## Q2

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def crop_grid(img, num_horizontal_grid, num_vertical_grid, line_color):
    """
    Divide an image into smaller patches using cropping and draw grid lines.
    
    Parameters:
        img (numpy.ndarray): The source image.
        num_horizontal_grid (int): Number of patches along the x-axis.
        num_vertical_grid (int): Number of patches along the y-axis.
        line_color (tuple): Color of the grid line in BGR format (e.g., (255, 0, 0) for blue).
    
    Returns:
        numpy.ndarray: Image with grid lines.
    """
    img_copy = img.copy()
    height, width = img_copy.shape[:2]
    
    # Calculate the size of each patch
    patch_width = width // num_horizontal_grid
    patch_height = height // num_vertical_grid
    
    # Draw vertical grid lines
    for i in range(1, num_horizontal_grid):
        x = i * patch_width
        cv2.line(img_copy, (x, 0), (x, height), line_color, 1)
    
    # Draw horizontal grid lines
    for i in range(1, num_vertical_grid):
        y = i * patch_height
        cv2.line(img_copy, (0, y), (width, y), line_color, 1)
    
    return img_copy

# Load the specified image
img = cv2.imread("Downloads/images/images/dog.jfif")

# Define grid parameters
num_horizontal_grid = 4
num_vertical_grid = 4
line_color = (255, 0, 0)  # Blue color in BGR

# Apply the crop_grid function
output_img_with_grid = crop_grid(img, num_horizontal_grid, num_vertical_grid, line_color)

# Display the image with grid
plt.imshow(cv2.cvtColor(output_img_with_grid, cv2.COLOR_BGR2RGB))
plt.title('Image with Grid')
plt.axis('off')
plt.show()
```

## Q3

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def change_brightness_opencv(image, value):
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    
    hsv[:, :, 2] = cv2.add(hsv[:, :, 2], value)
    
    bright_image = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)
    return bright_image
    
img = cv2.imread("Downloads/images/images/dog.jfif")

bright_img = change_brightness_opencv(img, 50)  


plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(bright_img, cv2.COLOR_BGR2RGB))
plt.title('Brightness Increased')
plt.axis('off')

plt.show()
```
